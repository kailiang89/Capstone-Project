# Content based using categories

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(restaurants_in_MT['categories'])
tfidf_matrix.shape
```

```python
from  sklearn.metrics.pairwise  import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Build a 1-dimensional array with restaurant names
titles = restaurants_in_MT['business_id']
indices = pd.Series(restaurants_in_MT.index, index=restaurants_in_MT['business_id'])

# Function that get restuarant recommendations based on the cosine similarity score of restaurant categories
def genre_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]
    movie_indices = [i[0] for i in sim_scores]
    return restaurants_in_MT.iloc[movie_indices][['name','stars','categories']]
```

# Reading and Merging new dataframes for topic modeling Content based

```python
# merged the restaurants and reviews to a new dataset 
restaurants_df_combined = pd.merge(yelp_review_df, restaurants_df, how='left', on=['business_id'])
```

```python
# reviews_MT are the reviews given by user to the restuarants in Montreal  
reviews_MT = restaurants_df_combined[restaurants_df_combined.city == 'Montréal']
reviews_MT.reset_index(inplace=True)
reviews_MT.drop('index',axis =1)
```

```python
# a dataframe based on the groupby of restaurants and the sum of all the reviews.
comb_text_df = reviews_MT.groupby('business_id').apply(lambda x: x.sum())
```

```python
# added a new column called comb_text to slowly append the edited topics into the dataframe 
comb_text_df['comb_text'] = '0'
comb_text_df
```

```python
comb_text_df['text']
```

# Topic modeling

```python
def display_topics(model, feature_names, no_top_words):
    list_of_topics = []
    for topic_idx, topic in enumerate(model.components_):
        list_of_topics.append(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))
    return list_of_topics
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation

for i in range(len(comb_text_df)): 
    document1 = comb_text_df['text'][i]
    document1 = document1.split('.')
 
    tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=1, max_features=1000, stop_words='english', ngram_range=(1,3))
    tfidf = tfidf_vectorizer.fit_transform(document1)
    tfidf_feature_names = tfidf_vectorizer.get_feature_names()
    
    # Run NMF
    nmf = NMF(n_components=3, random_state=1, alpha=.1, l1_ratio=.5, init='nndsvd').fit(tfidf)
    l = display_topics(nmf, tfidf_feature_names, 5)
    l2 = []
    for x in l:
        l2.append(x.encode("utf-8"))
    comb_text_df['comb_text'][i] = ''.join(l2)
    
    
```
